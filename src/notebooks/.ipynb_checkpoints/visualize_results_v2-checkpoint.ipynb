```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset_df = pd.read_csv('/Users/amuhebwa/Documents/Stanford/Research_Code/CarbonMeasure_Responsibility/src/results/merged_metrics.csv')
```

```python
dataset_df['model_id'] = dataset_df['model_id'].replace({'swinv2': 'swin', 'convnext': 'ConvNext', 'deit':'DeiT', 'vit': 'ViT'})
#dataset_df['accuracy'] = dataset_df['accuracy']*100
dataset_df['experiment_name'] = dataset_df['experiment_name'].replace('PublicHarvestNet', 'HarvestNet')
```

```python
tiny_models_df = dataset_df[dataset_df['model_size'] == 'Tiny'].reset_index(drop=True)
small_models_df = dataset_df[dataset_df['model_size'] == 'Small'].reset_index(drop=True)
experiment_names = dataset_df['experiment_name'].unique()
```

```python
dataset_df
```

```python
# Define the color palette
plt.clf()
palette = sns.color_palette("colorblind", 4)

# Create subplots with 3 rows and 3 columns for MCC, Energy Consumed, and Carbon Emissions
# fig, axes = plt.subplots(3, 3, figsize=(18, 18))
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
#fig, axes = plt.subplots(3, 3, figsize=(10, 10))

# Plot grouped bar plots for MCC, Energy Consumed, and Carbon Emissions for each experiment
for idx, experiment in enumerate(experiment_names):
    experiment_data = dataset_df[dataset_df['experiment_name'] == experiment]
    
    # Combined data for MCC
    combined_data_mcc = experiment_data.pivot(index='model_id', columns='model_size', values='mcc').reset_index()
    
    # Plotting MCC
    ax_mcc = axes[0, idx]
    bar_width = 0.35
    bar1 = np.arange(len(combined_data_mcc['model_id']))
    bar2 = [x + bar_width for x in bar1]
    
    ax_mcc.bar(bar1, combined_data_mcc['Tiny'], color=palette[2], width=bar_width, edgecolor='k', label='Tiny')
    ax_mcc.bar(bar2, combined_data_mcc['Small'], color=palette[3], width=bar_width, edgecolor='k', label='Small')
    
    # ax_mcc.set_title(f'{experiment} - MCC')
    ax_mcc.set_title(f'{experiment}')
    ax_mcc.set_xticks([r + bar_width for r in range(len(combined_data_mcc['model_id']))])
    ax_mcc.set_xticklabels(combined_data_mcc['model_id'])
    ax_mcc.set_ylabel('MCC')
    ax_mcc.set_ylim(0, 1)
    ax_mcc.legend()
    ax_mcc.grid(True, color='gray', linewidth=0.1, alpha=0.7)
    sns.despine()

    # Combined data for Energy Consumed
    combined_data_energy = experiment_data.pivot(index='model_id', columns='model_size', values='energy_consumed').reset_index()
    
    # Plotting Energy Consumed
    ax_energy = axes[1, idx]
    bar1 = np.arange(len(combined_data_energy['model_id']))
    bar2 = [x + bar_width for x in bar1]
    
    ax_energy.bar(bar1, combined_data_energy['Tiny'], color=palette[2], width=bar_width, edgecolor='k', label='Tiny')
    ax_energy.bar(bar2, combined_data_energy['Small'], color=palette[3], width=bar_width, edgecolor='k', label='Small')
    
    # ax_energy.set_title(f'{experiment} - Energy Consumed')
    ax_energy.set_xticks([r + bar_width for r in range(len(combined_data_energy['model_id']))])
    ax_energy.set_xticklabels(combined_data_energy['model_id'])
    ax_energy.set_ylabel('Energy Consumed (kWh)')
    ax_energy.legend()
    ax_energy.grid(True, color='gray', linewidth=0.1, alpha=0.7)
    sns.despine()

    # Combined data for Carbon Emissions
    combined_data_carbon = experiment_data.pivot(index='model_id', columns='model_size', values='carbon_emissions').reset_index()
    
    # Plotting Carbon Emissions
    ax_carbon = axes[2, idx]
    bar1 = np.arange(len(combined_data_carbon['model_id']))
    bar2 = [x + bar_width for x in bar1]
    
    ax_carbon.bar(bar1, combined_data_carbon['Tiny'], color=palette[2], width=bar_width, edgecolor='k', label='Tiny')
    ax_carbon.bar(bar2, combined_data_carbon['Small'], color=palette[3], width=bar_width, edgecolor='k', label='Small')
    
    # ax_carbon.set_title(f'{experiment} - Carbon Emissions')
    ax_carbon.set_xticks([r + bar_width for r in range(len(combined_data_carbon['model_id']))])
    ax_carbon.set_xticklabels(combined_data_carbon['model_id'])
    ax_carbon.set_ylabel('Carbon Emissions (kg CO2)')
    ax_carbon.legend()
    ax_carbon.grid(True, color='gray', linewidth=0.1, alpha=0.7)
    sns.despine()

# Adjust layout
plt.tight_layout()
#plt.show()
plt.savefig('/Users/amuhebwa/Documents/Stanford/Research_Code/CarbonMeasure_Responsibility/src/results/graphs/mcc_experiments_comparison.png')
```
